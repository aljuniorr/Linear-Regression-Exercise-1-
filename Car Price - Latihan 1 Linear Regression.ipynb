```python
# Library
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load Data
df = pd.read_csv('CarPrice_Assignment.csv')
pd.set_option('display.max_columns', 50)
df
```

```python
# Columns and Value Check
list = []
for i in df.columns:
    list.append([i, df[i].unique()])

pd.DataFrame(list, columns=['Columns Name','Unique Value'])
```

```python
# Check Missing Value
df.isna().sum()
```

```python
# Splitting Company Name
CompanyName = df['CarName'].apply(lambda x : x.split(' ')[0])
df.insert(3,"CompanyName",CompanyName)
df.drop(['CarName'],axis=1,inplace=True)
df.head()
```

```python
# Check Unique Value on CompanyName
df.CompanyName.unique()
```

```python
# Rename Invalid Brand Name
df.CompanyName = df.CompanyName.str.lower()

def replace_name(a,b):
    df.CompanyName.replace(a,b,inplace=True)

replace_name('maxda','mazda')
replace_name('porcshce','porsche')
replace_name('toyouta','toyota')
replace_name('vokswagen','volkswagen')
replace_name('vw','volkswagen')

df.CompanyName.unique()
```

```python
# Create DF Prediction
hp_pred = df[['price','horsepower']]
hp_pred.head()
```

```python
# Regression Model
reg_model = sm.OLS(hp_pred['horsepower'], sm.add_constant(hp_pred[['price']]))
hasil = reg_model.fit()
print(hasil.summary())
```
